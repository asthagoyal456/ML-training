```python
import pandas as pd
import warnings
import tkinter as tk
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
app = tk.Tk()
app.title('House Price Predictor')
app.geometry('500x400')
app.configure(bg = '#088778')

     
income = tk.StringVar()
house_age = tk.StringVar()
rooms = tk.StringVar()
bedrooms = tk.StringVar()
population = tk.StringVar()
    
tk.Label(app, text='Avg. Area Income:',bg='#088778',font=('Times New Roman', 12, 'bold')).place(x=25, y=35)
tk.Label(app, text='Avg. Area House Age:',bg='#088778',font=('Times New Roman', 12, 'bold')).place(x=25, y=70)
tk.Label(app, text='Avg. Area Number of Rooms:',bg='#088778',font=('Times New Roman', 12, 'bold')).place(x=25, y=105)
tk.Label(app, text='Avg. Area Number of Bedrooms:',bg='#088778',font=('Times New Roman', 12, 'bold')).place(x=25, y=140)
tk.Label(app, text='Area Population:',bg='#088778',font=('Times New Roman', 12, 'bold')).place(x=25, y=175)



tk.Entry(app, width=30, bg='#ffffff', textvariable=income).place(x=270,y=35)
tk.Entry(app, width=30, bg='#ffffff', textvariable=house_age).place(x=270,y=70)
tk.Entry(app, width=30, bg='#ffffff', textvariable=rooms).place(x=270,y=105)
tk.Entry(app, width=30, bg='#ffffff', textvariable=bedrooms).place(x=270,y=140)
tk.Entry(app, width=30, bg='#ffffff', textvariable=population).place(x=270,y=175)

df=pd.read_csv('housing.csv')
x=df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms' , 'Avg. Area Number of Bedrooms', 'Area Population']]
y=df['Price']
xtrain , xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=10)
model=LinearRegression()
model.fit(xtrain,ytrain)
model.coef_
pd.DataFrame(model.coef_,columns=['coef_'],index=x.columns)
ypred=model.predict(xtest)
final_model=model
pd.to_pickle(final_model,'HousePricePredictor')
model=pd.read_pickle('HousePricePredictor')

def predictor():
    
    a=income.get()
    b=house_age.get()
    c=rooms.get()
    d=bedrooms.get()
    e=population.get()
    
    income.set('')
    house_age.set('')
    rooms.set('')
    bedrooms.set('')
    population.set('')
    query = pd.DataFrame({'Avg. Area Income':[a],
                     'Avg. Area House Age':[b],
                      'Avg. Area Number of Rooms':[c],
                      'Avg. Area Number of Bedrooms':[d],
                      'Area Population':[e]
                     })
    Predictor_var=tk.Variable(app)
    Predictor_var.set('%.1f$'%(model.predict(query)[0]))
   
    
    Predict_the_price=tk.Label(app,text='Predicted Price :',bg='#088778',font=('Times New Roman', 12, 'bold')).place(x=25,y=300)
    Predict_the_price=tk.Label(app,text ="Predicted Price :",textvariable=Predictor_var,bg='#088778',font=('Times New Roman', 12, 'bold')).place(x=270,y=300)
    
tk.Button(app, text='Suggest Price', command=predictor , bg='#3fa67d').place(x=180, y=240)

app.mainloop()
```
